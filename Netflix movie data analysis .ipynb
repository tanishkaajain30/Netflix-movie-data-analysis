categorizing Vote_Average column

```python
# we would cut the Vote_Average values and make 4 categories: popular, average, below_avg, not_pouplar yo describe it more using catigorize_col() 
# function provided above
```

```python
# user defined function --> 
def catigorize_col(df, col, labels):
    
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    
    df[col] = pd.cut(df[col], edges , labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ['not_pouplar', 'below_avg', 'average', 'popular']

catigorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
